```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from pyEDM import CCM
```

```python
motor_ccm_df = pd.read_csv("../data/motor_ccm_df.csv")

motor_res = CCM(
    dataFrame=motor_ccm_df,
    E=6,
    sample=100,
    embedded=True,
    columns=[f'motor_eeg_{i+1}' for i in range(6)],
    target="acc",
    showPlot=True,
    returnObject=True,
    libSizes="100 150 5",
);
```
